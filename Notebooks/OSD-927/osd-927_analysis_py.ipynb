#### Method validation
- GemCode Single-Cell 3’ Gel Bead and Library v3
- Read 1 structure: 28bp total
- Barcode length: 16bp
- UMI length: 12bp

```python
# 10x v3 = 16bp Barcode, 12bp UMI
# 10x v2 = 16bp Barcode, 10bp UMI
bc_len = 16
umi_len = 12
```

```python
## Validate scRNA-Seq subsample 1
# Dictionaries
cell_umis = collections.defaultdict(set)
read_counts = collections.defaultdict(int)

with open('r1_sub.fastq', 'r') as f:
    for i, line in enumerate(f):
        if i % 4 == 1:
            barcode = line[:16]
            umi = line[16:28]
            cell_umis[barcode].add(umi)
            read_counts[barcode] += 1

# Saturation (unique reads)
top_cell = max(read_counts, key=read_counts.get)
unique_mols = len(cell_umis[top_cell])
total_reads = read_counts[top_cell]

saturation = 1 - (unique_mols / total_reads)
print(f"Top Cell ({top_cell}) Analysis:")
print(f"- Reads: {total_reads}")
print(f"- Unique Molecules (UMIs): {unique_mols}")
print(f"- Saturation: {saturation:.2%}")
```

```python
## Validate scRNA-Seq subsample 2 
# Count sequences
r2_counts = collections.Counter()
with open('r2_sub.fastq', 'r') as f:
    for i, line in enumerate(f):
        if i % 4 == 1:
            r2_counts[line.strip()] += 1

print("Top 5 overrepresented sequences:")
for seq, count in r2_counts.most_common(5):
    percentage = (count / 100000) * 100
    print(f"    [{count} reads | {percentage:.2f}%]: {seq}")
```
